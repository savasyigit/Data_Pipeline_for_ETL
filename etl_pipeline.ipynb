# Building a Reusable Data Extraction Function

## 🇬🇧 English

The previous approach successfully loads the datasets, but it contains repetitive code. One of the most important software development principles is **DRY (Don't Repeat Yourself)**, which encourages developers to avoid writing the same logic multiple times.

Instead of repeating the same code for every dataset, we can create a reusable function that performs the extraction process. This approach improves code readability, maintainability, and scalability, especially when working with multiple data sources.

In this section, we will create a function named **`extract()`** that automates the data extraction process.

The function will perform the following tasks:

1. Accept the file path as an input parameter (`file_path`).
2. Read the CSV file into a Pandas DataFrame.
3. Display useful information about the dataset, including:
   - Number of rows and columns
   - Data types of each column
4. Print instructions explaining how to use the returned DataFrame.
5. Return the extracted DataFrame for use in subsequent ETL stages.

After defining the function, it will be executed twice:

- Once for **apps_data.csv**
- Once for **review_data.csv**

Finally, one of the extracted DataFrames will be displayed to verify that the extraction process completed successfully.

---

## 🇹🇷 Türkçe

Önceki yöntemde veri setleri başarıyla okunmaktadır; ancak aynı kod blokları birden fazla kez tekrar edilmektedir. Yazılım geliştirmede en önemli prensiplerden biri olan **DRY (Don't Repeat Yourself - Kendini Tekrar Etme)** ilkesi, aynı kodun tekrar tekrar yazılmasını önermektedir.

Bu nedenle her veri seti için aynı işlemleri yeniden yazmak yerine, veri okuma işlemini gerçekleştiren yeniden kullanılabilir bir fonksiyon oluşturacağız. Bu yaklaşım kodun okunabilirliğini artırır, bakımını kolaylaştırır ve farklı veri kaynaklarıyla çalışırken ölçeklenebilir bir yapı sağlar.

Bu bölümde **`extract()`** isimli bir fonksiyon geliştirilecektir.

Fonksiyon aşağıdaki işlemleri gerçekleştirecektir:

1. Dosya yolunu (`file_path`) parametre olarak alacaktır.
2. CSV dosyasını bir Pandas DataFrame'ine okuyacaktır.
3. Veri setine ait temel bilgileri ekrana yazdıracaktır:
   - Satır ve sütun sayısı
   - Her sütunun veri tipi
4. Fonksiyonun döndürdüğü DataFrame'in nasıl kullanılacağına dair bilgilendirici mesajlar gösterecektir.
5. Okunan DataFrame'i ETL sürecinin sonraki aşamalarında kullanılmak üzere döndürecektir.

Fonksiyon tanımlandıktan sonra iki farklı veri seti için çalıştırılacaktır:

- **apps_data.csv**
- **review_data.csv**

Son olarak, okunan DataFrame'lerden biri görüntülenerek veri çıkarma işleminin başarıyla tamamlandığı doğrulanacaktır.

```python
# Define a reusable function to load a CSV file and display basic information.
# This function improves code reusability by eliminating repetitive code.
# Bir CSV dosyasını okuyup temel bilgilerini görüntüleyen yeniden kullanılabilir bir fonksiyon tanımlanır.
# Bu yaklaşım, aynı işlemleri tekrar yazmayı önleyerek kodun daha düzenli ve sürdürülebilir olmasını sağlar.

def extract(file_path):
    """
    Description:
        This function reads a CSV file and displays basic information
        about the extracted dataset.

    Açıklama:
        Bu fonksiyon belirtilen yoldaki CSV dosyasını okur ve
        veri setine ait temel bilgileri ekrana yazdırır.

    Args:
        file_path (str): Path to the CSV file.

    Parametreler:
        file_path (str): Okunacak CSV dosyasının yolu.

    Returns:
        pd.DataFrame: The loaded dataset.

    Döndürür:
        pd.DataFrame: Okunan veri setini DataFrame olarak döndürür.
    """

    # Read the CSV file into a Pandas DataFrame.
    # CSV dosyası Pandas DataFrame'ine aktarılır.
    data = pd.read_csv(file_path)

    # Display general information about the dataset.
    # Veri seti hakkında genel bilgiler görüntülenir.

    print(f"Here is a little bit of information about the data stored in '{file_path}':")
    print(f"'{file_path}' dosyasında bulunan veri setine ait temel bilgiler aşağıda gösterilmektedir.\n")

    print(f"There are {data.shape[0]} rows and {data.shape[1]} columns in this DataFrame.")
    print(f"Bu DataFrame toplam {data.shape[0]} satır ve {data.shape[1]} sütundan oluşmaktadır.\n")

    print("The columns in this DataFrame have the following data types:")
    print("Bu DataFrame'deki sütunların veri tipleri aşağıdaki gibidir:")
    print(data.dtypes)

    print(f"\nTo view the DataFrame extracted from '{file_path}', display the value returned by this function!")
    print(f"'{file_path}' dosyasından okunan veri setini görüntülemek için fonksiyonun döndürdüğü değeri görüntüleyebilirsiniz.\n")

    # Return the DataFrame for further analysis.
    # DataFrame daha sonraki analizlerde kullanılmak üzere döndürülür.
    return data


# Load the applications dataset using the reusable function.
# Yeniden kullanılabilir fonksiyon kullanılarak uygulama veri seti yüklenir.
apps_data = extract("apps_data.csv")

# Load the reviews dataset using the same function.
# Aynı fonksiyon kullanılarak inceleme veri seti yüklenir.
reviews_data = extract("review_data.csv")

# Display one of the extracted DataFrames for inspection.
# Okunan veri setlerinden biri görüntülenerek yapısı incelenir.
reviews_data
```

# Transforming Data

## 🇬🇧 English

The **Transform** stage is the second phase of the ETL (Extract, Transform, Load) process. During this stage, the extracted data is cleaned, filtered, combined, and prepared for analysis. Data transformation improves data quality and ensures that only relevant information is retained for downstream processes.

In this notebook, we are interested in analyzing applications that belong to the **"FOOD_AND_DRINK"** category. To accomplish this, we will create a reusable **`transform()`** function that performs several data transformation operations.

The function will complete the following tasks:

1. Accept the following parameters:
   - `apps`
   - `reviews`
   - `category`
   - `min_rating`
   - `min_reviews`

2. Remove duplicate records from both the applications and reviews datasets.

3. Filter only the applications that belong to the selected category.

4. Retrieve the corresponding reviews for those applications and calculate the **average sentiment polarity** for each application.

5. Merge the aggregated review information back into the applications dataset.

6. Keep only the columns required for the analysis:

   - `App`
   - `Rating`
   - `Reviews`
   - `Installs`
   - `Sentiment_Polarity`

7. Convert the **Reviews** column into an integer data type.

8. Filter applications that satisfy both conditions:
   - Rating greater than the specified minimum rating.
   - Number of reviews greater than the specified minimum review count.

9. Sort the final dataset by:
   - Rating (descending)
   - Number of reviews (descending)

10. Save the transformed dataset as **top_apps.csv** for later use in the loading stage.

Finally, the function will be executed for the **"FOOD_AND_DRINK"** category using:
- Minimum rating: **4.0**
- Minimum number of reviews: **1000**

The resulting dataset will contain only the highest-rated and most frequently reviewed applications that satisfy the specified criteria.

---

## 🇹🇷 Türkçe

**Transform (Dönüştürme)** aşaması, ETL (Extract, Transform, Load) sürecinin ikinci adımıdır. Bu aşamada veri temizlenir, filtrelenir, birleştirilir ve analiz için uygun hale getirilir. Veri dönüştürme işlemleri, veri kalitesini artırırken yalnızca analiz için gerekli bilgilerin korunmasını sağlar.

Bu notebook'ta **"FOOD_AND_DRINK"** kategorisindeki uygulamalar analiz edilecektir. Bunun için çeşitli veri dönüştürme işlemlerini gerçekleştiren yeniden kullanılabilir **`transform()`** fonksiyonu oluşturulacaktır.

Fonksiyon aşağıdaki işlemleri gerçekleştirecektir:

1. Aşağıdaki parametreleri alacaktır:
   - `apps`
   - `reviews`
   - `category`
   - `min_rating`
   - `min_reviews`

2. Hem uygulama hem de inceleme veri setlerindeki yinelenen kayıtları kaldıracaktır.

3. Yalnızca seçilen kategoriye ait uygulamaları filtreleyecektir.

4. Bu uygulamalara ait kullanıcı yorumlarını alacak ve her uygulama için **ortalama duygu (Sentiment Polarity)** skorunu hesaplayacaktır.

5. Hesaplanan duygu skorlarını uygulama veri seti ile birleştirecektir.

6. Analizde kullanılacak aşağıdaki sütunları koruyacaktır:

   - `App`
   - `Rating`
   - `Reviews`
   - `Installs`
   - `Sentiment_Polarity`

7. **Reviews** sütununu tam sayı veri tipine dönüştürecektir.

8. Aşağıdaki kriterleri sağlayan uygulamaları filtreleyecektir:
   - Belirlenen minimum puandan daha yüksek puana sahip olmak.
   - Belirlenen minimum inceleme sayısından daha fazla incelemeye sahip olmak.

9. Sonuç veri setini aşağıdaki ölçütlere göre azalan sırada sıralayacaktır:
   - Uygulama puanı (Rating)
   - İnceleme sayısı (Reviews)

10. Dönüştürülen veri setini, ETL sürecinin **Load** aşamasında kullanılmak üzere **top_apps.csv** dosyası olarak kaydedecektir.

Son olarak fonksiyon aşağıdaki parametrelerle çalıştırılacaktır:

- Kategori: **"FOOD_AND_DRINK"**
- Minimum puan: **4.0**
- Minimum inceleme sayısı: **1000**

Elde edilen veri seti, belirtilen kriterleri karşılayan en yüksek puanlı ve en fazla kullanıcı incelemesine sahip uygulamalardan oluşacaktır. Bu veri seti bir sonraki aşamada **PostgreSQL** veritabanına aktarılacaktır.

```python
# Define a reusable function to clean, transform, and filter the applications dataset.
# The function combines application and review data, applies several transformations,
# and returns the most popular applications based on the specified criteria.
#
# Uygulama ve inceleme veri setlerini temizleyen, dönüştüren ve filtreleyen
# yeniden kullanılabilir bir fonksiyon tanımlanır. Fonksiyon, iki veri setini
# birleştirerek belirlenen kriterlere göre en popüler uygulamaları döndürür.

def transform(apps, reviews, category, min_rating, min_reviews):
    """
    Description:
        Transform the apps and reviews data to include only applications in the
        specified category, calculate the average sentiment polarity, keep the
        required columns, filter less popular applications, and return the
        transformed DataFrame sorted by rating and number of reviews.

    Açıklama:
        Uygulama ve inceleme verilerini dönüştürerek yalnızca belirtilen
        kategoriye ait uygulamaları seçer, ortalama duygu (sentiment)
        skorunu hesaplar, gerekli sütunları korur, popüler olmayan
        uygulamaları filtreler ve sonucu puan ile inceleme sayısına göre
        sıralanmış şekilde döndürür.

    Args:
        apps (pd.DataFrame): Applications dataset.
        reviews (pd.DataFrame): Reviews dataset.
        category (str): Target application category.
        min_rating (float): Minimum application rating.
        min_reviews (int): Minimum number of reviews.

    Parametreler:
        apps (pd.DataFrame): Uygulama veri seti.
        reviews (pd.DataFrame): İnceleme veri seti.
        category (str): Filtrelenecek uygulama kategorisi.
        min_rating (float): Minimum uygulama puanı.
        min_reviews (int): Minimum inceleme sayısı.

    Returns:
        pd.DataFrame: Transformed and filtered dataset.

    Döndürür:
        pd.DataFrame: Dönüştürülmüş ve filtrelenmiş veri seti.
    """

    # Display information about the transformation process.
    # Gerçekleştirilecek veri dönüştürme işlemi hakkında bilgi verilir.
    print(
        f"Transforming data to curate a dataset with all {category} apps and their "
        f"corresponding reviews with a rating of at least {min_rating} and "
        f"{min_reviews} reviews."
    )

    print(
        f"Veri dönüştürme işlemi başlatıldı. '{category}' kategorisindeki, "
        f"puanı en az {min_rating} olan ve en az {min_reviews} incelemeye sahip "
        f"uygulamalar analiz edilmektedir.\n"
    )

    # Remove duplicate application records based on the App column.
    # App sütununa göre yinelenen uygulama kayıtları kaldırılır.
    apps = apps.drop_duplicates(["App"])

    # Remove duplicate review records.
    # Yinelenen inceleme kayıtları kaldırılır.
    reviews = reviews.drop_duplicates()

    # Filter applications that belong to the selected category.
    # Belirlenen kategoriye ait uygulamalar filtrelenir.
    subset_apps = apps.loc[apps["Category"] == category, :]

    # Keep only reviews that belong to the filtered applications.
    # Yalnızca filtrelenen uygulamalara ait incelemeler seçilir.
    subset_reviews = reviews.loc[
        reviews["App"].isin(subset_apps["App"]),
        ["App", "Sentiment_Polarity"]
    ]

    # Calculate the average sentiment polarity for each application.
    # Her uygulama için ortalama duygu (sentiment) skoru hesaplanır.
    aggregated_reviews = subset_reviews.groupby(by="App").mean()

    # Merge application information with the aggregated review scores.
    # Uygulama bilgileri ile ortalama duygu skorları birleştirilir.
    joined_apps_reviews = subset_apps.join(
        aggregated_reviews,
        on="App",
        how="left"
    )

    # Keep only the columns required for the analysis.
    # Analiz için gerekli sütunlar seçilir.
    filtered_apps_reviews = joined_apps_reviews.loc[
        :,
        ["App", "Rating", "Reviews", "Installs", "Sentiment_Polarity"]
    ]

    # Convert the Reviews column to integer format.
    # Reviews sütunu tam sayı veri tipine dönüştürülür.
    filtered_apps_reviews = filtered_apps_reviews.astype({"Reviews": "int32"})

    # Filter applications that satisfy the minimum rating and review thresholds.
    # Minimum puan ve minimum inceleme sayısı kriterlerini sağlayan uygulamalar seçilir.
    top_apps = filtered_apps_reviews.loc[
        (filtered_apps_reviews["Rating"] > min_rating)
        & (filtered_apps_reviews["Reviews"] > min_reviews),
        :
    ]

    # Sort the applications by rating and review count in descending order.
    # Uygulamalar puan ve inceleme sayısına göre azalan şekilde sıralanır.
    top_apps.sort_values(
        by=["Rating", "Reviews"],
        ascending=False,
        inplace=True
    )

    # Reset the index after sorting.
    # Sıralama sonrasında indeks yeniden oluşturulur.
    top_apps.reset_index(drop=True, inplace=True)

    # Save the transformed dataset as a CSV file.
    # Dönüştürülen veri seti CSV dosyası olarak kaydedilir.
    top_apps.to_csv("top_apps.csv", index=False)

    # Display information about the generated dataset.
    # Oluşturulan veri seti hakkında bilgi verilir.
    print(
        f"The transformed DataFrame contains {top_apps.shape[0]} rows "
        f"and {top_apps.shape[1]} columns. "
        f"It has been saved as 'top_apps.csv' and is ready for further analysis."
    )

    print(
        f"Dönüştürülen DataFrame {top_apps.shape[0]} satır ve "
        f"{top_apps.shape[1]} sütundan oluşmaktadır. "
        f"Veri seti 'top_apps.csv' dosyası olarak kaydedildi ve "
        f"analiz için hazır hale getirildi."
    )

    # Return the transformed DataFrame.
    # Dönüştürülen DataFrame döndürülür.
    return top_apps


# Execute the transformation function using the specified parameters.
# Belirlenen parametrelerle veri dönüştürme fonksiyonu çalıştırılır.
top_apps_data = transform(
    apps=apps_data,
    reviews=reviews_data,
    category="FOOD_AND_DRINK",
    min_rating=4.0,
    min_reviews=1000
)

# Display the transformed dataset.
# Dönüştürülen veri seti görüntülenir.
top_apps_data
```

# Loading Data

## 🇬🇧 English

The **Load** stage is the final step of the ETL (Extract, Transform, Load) process. After the data has been extracted and transformed, it is loaded into a database where it can be stored, queried, and analyzed efficiently.

In this notebook, instead of using SQLite, we will load the transformed dataset into a **PostgreSQL** database. PostgreSQL is one of the most popular open-source relational database management systems and is widely used in production-grade data engineering projects.

To accomplish this, we will use **Pandas**, **SQLAlchemy**, and the **psycopg2** PostgreSQL driver.

The **`load()`** function will perform the following tasks:

1. Accept the following parameters:
   - `dataframe`
   - `host`
   - `port`
   - `database_name`
   - `user`
   - `password`
   - `table_name`

2. Create a connection to the PostgreSQL database using SQLAlchemy.

3. Load the DataFrame into the specified PostgreSQL table.
   - Replace the table if it already exists.
   - Do not write the DataFrame index.

4. Read the table back from PostgreSQL using `pd.read_sql()`.

5. Validate the loading process by comparing the dimensions of the original and loaded DataFrames.

6. Display informative messages indicating whether the loading process completed successfully.

7. Return the DataFrame retrieved from PostgreSQL.

Finally, the function will be executed using the transformed **`top_apps_data`** DataFrame and will store the data in the **`top_apps`** table of the target PostgreSQL database.

This validation step helps ensure that the ETL pipeline has successfully transferred the transformed data into the database without data loss.

## 🇹🇷 Türkçe

**Load (Yükleme)** aşaması, ETL (Extract, Transform, Load) sürecinin son adımıdır. Veriler çıkarılıp dönüştürüldükten sonra, uzun süreli saklama, sorgulama ve analiz işlemleri için bir veritabanına aktarılır.

Bu notebook'ta SQLite yerine dönüştürülen veri seti **PostgreSQL** veritabanına yüklenecektir. PostgreSQL, açık kaynaklı, güçlü ve üretim ortamlarında yaygın olarak kullanılan ilişkisel bir veritabanı yönetim sistemidir.

Bu işlem için **Pandas**, **SQLAlchemy** ve **psycopg2** PostgreSQL sürücüsü kullanılacaktır.

Oluşturulacak **`load()`** fonksiyonu aşağıdaki işlemleri gerçekleştirecektir:

1. Aşağıdaki parametreleri alacaktır:
   - `dataframe`
   - `host`
   - `port`
   - `database_name`
   - `user`
   - `password`
   - `table_name`

2. SQLAlchemy kullanılarak PostgreSQL veritabanına bağlantı oluşturacaktır.

3. DataFrame'i belirtilen PostgreSQL tablosuna yazacaktır.
   - Tablo mevcutsa yeniden oluşturulacaktır (`replace`).
   - DataFrame indeks sütunu veritabanına yazılmayacaktır.

4. Yazılan tablo doğrulama amacıyla `pd.read_sql()` fonksiyonu kullanılarak PostgreSQL'den tekrar okunacaktır.

5. Orijinal DataFrame ile veritabanından okunan DataFrame'in satır ve sütun sayıları karşılaştırılarak veri yükleme işlemi doğrulanacaktır.

6. Veri yükleme işleminin başarılı veya başarısız olduğunu belirten bilgilendirici mesajlar ekrana yazdırılacaktır.

7. PostgreSQL veritabanından okunan doğrulanmış DataFrame döndürülecektir.

Son olarak fonksiyon, dönüştürülen **`top_apps_data`** veri seti kullanılarak çalıştırılacak ve veriler hedef PostgreSQL veritabanındaki **`top_apps`** tablosuna aktarılacaktır.

Bu doğrulama adımı sayesinde ETL veri hattının dönüştürülen verileri PostgreSQL veritabanına eksiksiz ve doğru şekilde aktardığı kontrol edilmiş olacaktır.

```python
# Import the required library for connecting to PostgreSQL.
# PostgreSQL veritabanına bağlanmak için gerekli kütüphane içe aktarılır.
from sqlalchemy import create_engine

# Define a reusable function to load a DataFrame into a PostgreSQL database
# and validate that the data has been loaded correctly.
#
# Bir DataFrame'i PostgreSQL veritabanına yükleyen ve
# yüklenen veriyi doğrulayan yeniden kullanılabilir bir fonksiyon tanımlanır.

def load(dataframe, host, port, database_name, user, password, table_name):
    """
    Description:
        Loads a DataFrame into a PostgreSQL database and validates
        that the loaded table matches the original DataFrame.

    Açıklama:
        Bir DataFrame'i PostgreSQL veritabanına yükler ve
        yüklenen tablonun orijinal veriyle aynı yapıya sahip olduğunu doğrular.

    Args:
        dataframe (pd.DataFrame): Data to be loaded.
        host (str): PostgreSQL server address.
        port (int): PostgreSQL server port.
        database_name (str): Database name.
        user (str): PostgreSQL username.
        password (str): PostgreSQL password.
        table_name (str): Target table name.

    Parametreler:
        dataframe (pd.DataFrame): Yüklenecek veri.
        host (str): PostgreSQL sunucu adresi.
        port (int): PostgreSQL port numarası.
        database_name (str): Veritabanı adı.
        user (str): PostgreSQL kullanıcı adı.
        password (str): PostgreSQL şifresi.
        table_name (str): Verinin yazılacağı tablo adı.

    Returns:
        pd.DataFrame

    Döndürür:
        pd.DataFrame: Veritabanından okunarak doğrulanan veri.
    """

    # Create the PostgreSQL connection engine.
    # PostgreSQL bağlantısı oluşturulur.
    engine = create_engine(
        f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database_name}"
    )

    # Load the DataFrame into PostgreSQL.
    # DataFrame PostgreSQL tablosuna yazılır.
    dataframe.to_sql(
        name=table_name,
        con=engine,
        if_exists="replace",
        index=False
    )

    print("The original DataFrame has been successfully loaded into PostgreSQL.")
    print("Orijinal DataFrame başarıyla PostgreSQL veritabanına yüklendi.\n")

    # Read the table back from PostgreSQL.
    # Doğrulama amacıyla tablo tekrar okunur.
    loaded_dataframe = pd.read_sql(
        sql=f"SELECT * FROM {table_name}",
        con=engine
    )

    print("The loaded table has been read from PostgreSQL for validation.")
    print("Yüklenen tablo doğrulama amacıyla PostgreSQL veritabanından okunmuştur.\n")

    # Validate that both DataFrames have the same dimensions.
    # İki DataFrame'in aynı boyutta olup olmadığı doğrulanır.
    try:
        assert dataframe.shape == loaded_dataframe.shape

        print(
            f"Success! The '{table_name}' table has been successfully loaded and validated."
        )
        print(
            f"Başarılı! '{table_name}' tablosu PostgreSQL veritabanına başarıyla yüklendi ve doğrulandı."
        )

    except AssertionError:

        print(
            "Validation failed! The DataFrame shape differs before and after loading."
        )
        print(
            "Doğrulama başarısız! Veri yükleme öncesi ve sonrası DataFrame boyutları farklıdır."
        )

    return loaded_dataframe


# Load the transformed dataset into PostgreSQL.
# Dönüştürülen veri PostgreSQL veritabanına yüklenir.
top_apps_loaded = load(
    dataframe=top_apps_data,
    host="localhost",
    port=5432,
    database_name="market_research",
    user="postgres",
    password="12345",
    table_name="top_apps"
)

# Display the loaded data.
# Veritabanından okunan doğrulanmış veri görüntülenir.
top_apps_loaded
```
